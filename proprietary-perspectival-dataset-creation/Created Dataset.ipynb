## Import Data

```python
# libaries
import pandas as pd
```

### Charles's Sample Dataset

```python
music_sample_data = pd.read_csv("music_sample.csv")
music_sample_data
```

```python
# Loading the dataset again 
music_sample_data = pd.read_csv("music_sample.csv")

# Renaming the column to 'Track'
music_sample_data.rename(columns={"Track Name": "Track"}, inplace=True)
music_sample_data
```

### The dataset found by Weiting

Source: https://www.kaggle.com/datasets/thebumpkin/300-world-music-tracks-with-spotify-data

```python
world_hits = pd.read_csv("WorldHits.csv")
world_hits
```

### Combine Dataset from the above two

```python
# # Combining the datasets with important music features
common_columns = list(set(music_sample_data.columns) & set(world_hits.columns))
combined_data = pd.concat(
    [music_sample_data[common_columns], world_hits[common_columns]], 
    ignore_index=True
)
# Display the combined dataset
df_new = combined_data
df_new
```

### Short Conclusion

The new combined datasets music_sample_data and world_hits have been combined based on their shared columns, including Popularity, Tempo, Valence, Danceability, Loudness, and Energy. This unified dataset provides a comprehensive view of musical characteristics and popularity metrics, which enables detailed analysis of trends and patterns in music from 1992 to 2018. 

### After midterm report, our group found Billboard_Hot100_Songs_Spotify_1946-2022, so we also decided to combine with this new combined dataset. 

Source: https://www.kaggle.com/datasets/dhruvildave/billboard-the-hot-100-songs

```python
# billboard dataset
billboard_music = pd.read_csv("Billboard_Hot100_Songs_Spotify_1946-2022.csv")
billboard_music
```

```python
# change the column name from Song to Track
if 'Song' in billboard_music.columns:
    billboard_music.rename(columns={"Song": "Track"}, inplace=True)
```

```python
# Finding common columns between the existing combined dataset and `billboard_music`
common_columns_with_billboard = list(set(df_new.columns) & set(billboard_music.columns))

# Combining the existing combined dataset with `billboard_music` based on common columns
final_combined_data = pd.concat(
    [df_new[common_columns_with_billboard], billboard_music[common_columns_with_billboard]],
    ignore_index=True
)
# Display the first few rows of the final combined dataset
final_combined_data
```

### Conclusion

The final_combined_data is the dataset we mentioned during the demo day. It integrates key musical features from Charles's dataset, the dataset I found (on Kaggle), and Billboard dataset (on Kaggle). Spanning the time period from 1946 to 2022, this version offers a comprehensive and unified foundation for analyzing trends, patterns, and relationships in music data across significant societal eras. By including attributes such as Popularity, Tempo, Valence, Danceability, Loudness, and Energy, this dataset enables in-depth exploration of how musical characteristics have evolved and how they correlate with cultural and historical shifts.
