<h1>Computational method: Pandas, EDA, seaborn, and matplotlib

## Import Data

```python
# libaries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the dataset
df = pd.read_csv("Billboard_Hot100_Songs_Spotify_1946-2022.csv")
df
```

## Data Cleaning

```python
# no missing value found
df.dropna()
```

## Data Exploration with EDA

```python
# Filter rows where 'Artist(s) Genres' contains "pop"
pop_songs = df[df['Artist(s) Genres'].str.contains('pop', case=False, na=False)]
pop_songs
```

```python
# Before economic recession (2004 to 2006)
# Explore pop music between 2004 to 2006
df_before_recession = pop_songs[(pop_songs['Hot100 Ranking Year'] >= 2004) & (df['Hot100 Ranking Year'] <= 2006)]
pop_songs_before_recession = df_before_recession.shape[0]  
print(f"The number of pop songs before economic recession: {pop_songs_before_recession}")
```

```python
# The Great Recession was a global economic downturn that began in December 2007 and ended in June 2009
# Explore pop music between 2007 to 2009
df_recession = pop_songs[(pop_songs['Hot100 Ranking Year'] >= 2007) & (df['Hot100 Ranking Year'] <= 2009)]
pop_songs_during_recession = df_recession.shape[0]  
print(f"The number of pop songs duing economic recession: {pop_songs_during_recession}")
```

```python
# Explore pop music after recession between 2010 to 2012
df_after_recession = pop_songs[(pop_songs['Hot100 Ranking Year'] >= 2010) & (df['Hot100 Ranking Year'] <= 2012)]
pop_songs_after_recession = df_after_recession.shape[0]  
print(f"The number of pop songs duing economic recession: {pop_songs_after_recession}")
```

```python
# Group by year to count songs per year
pop_songs_by_year = pop_songs.groupby('Hot100 Ranking Year').size()

# Create a line plot
pop_songs_by_year.plot(kind='line', marker='o', figsize=(10, 6))
plt.axvspan(2004, 2006, color='blue', alpha=0.2, label='Before Recession')
plt.axvspan(2007, 2009, color='red', alpha=0.2, label='During Recession')
#plt.axvspan(2010, 2020, color='orange', alpha=0.2, label='After Recession')
plt.title('Number of Pop Songs Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.legend()
plt.show()
```

```python
# Filter and zoom in to include only years from 2000 to 2010
pop_songs_filtered = pop_songs[(pop_songs['Hot100 Ranking Year'] >= 2000) & (pop_songs['Hot100 Ranking Year'] <= 2010)]

# Group by year to count songs per year
pop_songs_by_year = pop_songs_filtered.groupby('Hot100 Ranking Year').size()

# Create a line plot
plt.figure(figsize=(10, 6))
pop_songs_by_year.plot(kind='line', marker='o')

# Highlight economic periods
plt.axvspan(2004, 2006, color='blue', alpha=0.2, label='Before Recession')
plt.axvspan(2007, 2009, color='red', alpha=0.2, label='During Recession')

# Add labels and legend
plt.title('Number of Pop Songs (2000–2010)')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.legend()
plt.show()
```

## Short Conclusion

The line plot illustrates the number of pop songs released annually between 2000 and 2010, highlighting key economic periods. The visualization demonstrates that the number of pop songs increased steadily during the recession period (2007–2009). Despite the economic challenges of the time, the trend highlights a rise in pop song production and popularity. This may suggest that pop music served as a form of entertainment and escapism, gaining momentum even during periods of economic downturn. The data indicates that the recession did not hinder the growth of pop music but may have contributed to its prominence as a cultural outlet.
